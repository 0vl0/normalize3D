## Normalizing the orientation of a 3D image with the numpy version

```python
import numpy as np
from inertia import normalize_orientation
from plot_field import plot_inertia_numpy
%matplotlib widget
```

Loading the 3D image

```python
image_3D = np.load("./npy/example1.npy")
plot_inertia_numpy(image_3D)
```

Normalizing the orientation <br>
Though this image is binary, the algorithm works with non binary images as well.

```python
image_3D_normalized = normalize_orientation(image_3D)
plot_inertia_numpy(image_3D_normalized)
```
